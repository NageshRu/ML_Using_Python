```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('housing_in_london_monthly_variables.csv')
df=df.head(5)

df
```

```python
'''Box plot '''
```

```python
df.boxplot(by='date',column=['houses_sold'],grid=False)
```

```python
sns.set_style('whitegrid')
```

```python
sns.boxplot(x='date',y='houses_sold',data=df)
```

```python
'''Histogram of csv data'''

df['houses_sold'].plot(kind='hist')
```

```python
'''Density plot of csv data'''


for airline in df:    
    sns.distplot(df['houses_sold'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = airline)
    
    
```

```python
'''Violin Plot'''

plt.violinplot(df['houses_sold'],vert=False)
```
